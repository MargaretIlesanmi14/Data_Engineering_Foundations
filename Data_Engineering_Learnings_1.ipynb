Working with Databases: Data engineers frequently move data between databases and data pipelines.

We'll start with `SQLite`, a simple built-in database that requires no setup.

`SQLite` is a lightweight, serverless database engine that stores data in a single file on your device. It supports standard SQL and is widely used for small applications, testing, and mobile development.


```python
import sqlite3

#Create an in-memory SQLite databse
conn = sqlite3.connect(":memory:")
cursor = conn.cursor()
```

```python
# Create a simple table
cursor.execute('''
    CREATE TABLE employees (
        id INTEGER PRIMARY KEY,
        name TEXT,
        department TEXT,
        salary INTEGER
    )
''')
```

```python
## insert into the table

cursor.executemany('''
    INSERT INTO employees (name, department, salary)
    VALUES (?, ?, ?)
''', [
    ('Alice', 'IT', 70000),
    ('Bob', 'HR', 50000),
    ('Charlie', 'Finance', 80000)
])
conn.commit()
```

```python
## querying the database

cursor.execute("SELECT * FROM employees WHERE salary > 70000")
rows = cursor.fetchall()
for row in rows:
    print(row)
```

# Data Pipelines

Data pipelines move and transform data from one place to another on a schedule or trigger. It helps you automate and manage the data flow process.

## Writing a Reusable Python ETL Script

ETL stands for Extract, Transform, Load.

Here’s a simple ETL pipeline that reads from a CSV, processes the data, and writes it back.



```python
import pandas as pd

def extract(file_path):
    return pd.read_csv(file_path)

def transform(df):
    # Convert all column names to lowercase
    df.columns = [col.upper() for col in df.columns]
    return df

def load(df, output_path):
    df.to_csv(output_path, index=False)
```

```python
# Reusable ETL function
def etl_pipeline(input_path, output_path):
    df = extract(input_path)
    df = transform(df)
    load(df, output_path)
```

```python
# Sample run
etl_pipeline("employees.csv", "clean_data.csv")
```

Logging and Error Handling: It’s important to log activities and handle errors in production pipelines.

You can use Python’s built-in `logging` module to track events in your ETL pipeline.

```python
import logging

# Configure logging
logging.basicConfig(level=logging.INFO)

def safe_etl_pipeline(input_path, output_path):

    try:
        logging.info("Starting ETL pipeline")
        df = extract(input_path)
        df = transform(df)
        load(df, output_path)
        logging.info("ETL pipeline completed successfully")

    except Exception as e:
        logging.error(f"Pipeline failed: {e}")

# Run safely
safe_etl_pipeline("employee.csv", "clean_data.csv")
```
